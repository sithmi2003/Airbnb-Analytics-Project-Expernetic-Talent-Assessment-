```python
listings["price"] = (
    listings["price"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

listings["price"] = pd.to_numeric(
    listings["price"],
    errors="coerce"
)
```

```python
print("Negative Prices")

print((listings["price"] <= 0).sum())
```

```python
print("Invalid Latitude")

print(
    ((listings["latitude"] < -90) |
    (listings["latitude"] > 90)).sum()
)
```

```python
print("Invalid Longitude")

print(
    ((listings["longitude"] < -180) |
    (listings["longitude"] > 180)).sum()
)
```
