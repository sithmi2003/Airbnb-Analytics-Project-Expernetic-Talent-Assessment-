```python
# ===============================================
# Airbnb Singapore Feature Engineering Pipeline
# ===============================================

import pandas as pd
import numpy as np


# ===============================================
# Load Master Dataset
# ===============================================

print("Loading master dataset...\n")

master = pd.read_csv(
    "D:/Talent assessment/Data/cleaned/master_dataset.csv"
)


print("Dataset loaded successfully")

print("Original Shape:", master.shape)



# ===============================================
# 1. Price Based Features
# ===============================================

print("\nCreating price features...")


# Price category

def price_category(price):

    if price < 100:
        return "Low"

    elif price < 300:
        return "Medium"

    else:
        return "High"



master["price_category"] = master["price"].apply(
    price_category
)



# Price per guest

master["price_per_guest"] = (
    master["price"] /
    master["accommodates"]
)



# Handle infinite values

master["price_per_guest"] = (
    master["price_per_guest"]
    .replace([np.inf, -np.inf], np.nan)
)



# ===============================================
# 2. Availability Features
# ===============================================

print("Creating availability features...")


# Availability percentage

master["availability_rate"] = (
    master["availability_365"] / 365
)



# Occupancy percentage

master["occupancy_rate"] = (
    (365 - master["availability_365"])
    / 365
)



# Availability category

def availability_category(value):

    if value < 100:
        return "Low Availability"

    elif value < 250:
        return "Medium Availability"

    else:
        return "High Availability"



master["availability_category"] = (
    master["availability_365"]
    .apply(availability_category)
)



# ===============================================
# 3. Host Features
# ===============================================

print("Creating host features...")


# Professional vs Individual host

master["host_type"] = master[
    "calculated_host_listings_count"
].apply(
    lambda x:
    "Professional Host"
    if x > 5
    else "Individual Host"
)



# Host listing category

def host_listing_category(value):

    if value == 1:
        return "Single Listing"

    elif value <= 5:
        return "Small Host"

    else:
        return "Professional Host"



master["host_listing_category"] = (
    master["calculated_host_listings_count"]
    .apply(host_listing_category)
)



# ===============================================
# 4. Review Features
# ===============================================

print("Creating review features...")


# Overall review score category

def review_category(score):

    if score >= 4.8:
        return "Excellent"

    elif score >= 4.5:
        return "Good"

    elif score >= 4:
        return "Average"

    else:
        return "Low"



master["review_category"] = (
    master["review_scores_rating"]
    .apply(review_category)
)



# High rated listing flag

master["high_rating"] = (
    master["review_scores_rating"] >= 4.8
)



# ===============================================
# 5. Property Features
# ===============================================

print("Creating property features...")


# Total sleeping capacity

master["total_capacity"] = (
    master["bedrooms"].fillna(0)
    +
    master["beds"].fillna(0)
)



# Property size category

def property_size(value):

    if value <= 2:
        return "Small"

    elif value <= 5:
        return "Medium"

    else:
        return "Large"



master["property_size"] = (
    master["accommodates"]
    .apply(property_size)
)



# ===============================================
# 6. Revenue Features
# ===============================================

print("Creating revenue features...")


if "estimated_revenue_l365d" in master.columns:


    master["revenue_per_booking"] = (
        master["estimated_revenue_l365d"]
        /
        master["number_of_reviews"]
    )


    master["revenue_per_booking"] = (
        master["revenue_per_booking"]
        .replace([np.inf, -np.inf], np.nan)
    )



# ===============================================
# 7. Date Features
# ===============================================

print("Creating date features...")


if "date" in master.columns:

    master["date"] = pd.to_datetime(
        master["date"]
    )


    master["year"] = (
        master["date"].dt.year
    )


    master["month"] = (
        master["date"].dt.month
    )


    master["day_of_week"] = (
        master["date"].dt.day_name()
    )



# ===============================================
# 8. Remove Duplicate Columns
# ===============================================

print("Removing duplicate columns...")

master = master.loc[
    :,
    ~master.columns.duplicated()
]



# ===============================================
# 9. Final Feature Summary
# ===============================================

print("\nFinal Dataset Information")
print("="*50)

print(master.shape)

print("\nNew Features Created:")

new_features = [
    "price_category",
    "price_per_guest",
    "availability_rate",
    "occupancy_rate",
    "availability_category",
    "host_type",
    "host_listing_category",
    "review_category",
    "high_rating",
    "total_capacity",
    "property_size"
]


for feature in new_features:

    if feature in master.columns:
        print("✓", feature)



# ===============================================
# Save Feature Dataset
# ===============================================

print("\nSaving feature dataset...")


master.to_csv(
    "D:/Talent assessment/Data/cleaned/master_features.csv",
    index=False
)


print("\nFeature engineering completed successfully!")
```
