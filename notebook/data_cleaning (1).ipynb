```python
import pandas as pd
import numpy as np
```

```python
print("Loading datasets...")

listings = pd.read_csv("D:/Talent assessment/Data/raw/listings.csv")
calendar = pd.read_csv("D:/Talent assessment/Data/raw/calendar.csv")
reviews = pd.read_csv("D:/Talent assessment/Data/raw/reviews.csv")
neighbourhoods = pd.read_csv("D:/Talent assessment/Data/raw/neighbourhoods.csv")

print("Datasets loaded successfully.\n")
```

```python
print("After loading:", listings.shape)
```

```python
print("Removing duplicate records...")

listings = listings.drop_duplicates()
calendar = calendar.drop_duplicates()
reviews = reviews.drop_duplicates()
neighbourhoods = neighbourhoods.drop_duplicates()
```

```python
print("After duplicates:", listings.shape)
```

```python
print("Removing empty columns...")

empty_columns = [
    "neighborhood_overview",
    "host_since",
    "host_response_time",
    "host_response_rate",
    "host_acceptance_rate",
    "host_thumbnail_url",
    "host_neighbourhood",
    "host_total_listings_count",
    "host_verifications",
    "neighbourhood",
    "calendar_updated",
    "instant_bookable"
]

listings.drop(columns=empty_columns, inplace=True, errors="ignore")
```

```python
print("After removing columns:", listings.shape)
```

```python
print("Handling missing values...")

# Text columns
listings["description"] = listings["description"].fillna("No Description")
listings["host_name"] = listings["host_name"].fillna("Unknown")
listings["bathrooms_text"] = listings["bathrooms_text"].fillna("Not Specified")

# Numeric columns
listings["bedrooms"] = listings["bedrooms"].fillna(
    listings["bedrooms"].median()
)

listings["beds"] = listings["beds"].fillna(
    listings["beds"].median()
)

# reviews_per_month
listings["reviews_per_month"] = listings["reviews_per_month"].fillna(0)
```

```python
print("Cleaning Reviews Dataset...")

# Remove duplicate rows
reviews = reviews.drop_duplicates()

# Fill missing reviewer names
reviews["reviewer_name"] = reviews["reviewer_name"].fillna("Unknown")

# Fill missing comments
reviews["comments"] = reviews["comments"].fillna("No Comment")

# Convert date
reviews["date"] = pd.to_datetime(
    reviews["date"],
    errors="coerce"
)

print("\nMissing values after cleaning:")
print(reviews.isnull().sum())
```

```python
print("After missing values:", listings.shape)
```

```python
print("Converting data types...")
# Convert price to numeric

listings["price"] = (
    listings["price"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

listings["price"] = pd.to_numeric(
    listings["price"],
    errors="coerce"
)

# Convert dates
calendar["date"] = pd.to_datetime(calendar["date"])

reviews["date"] = pd.to_datetime(reviews["date"])

# Convert availability
calendar["available"] = calendar["available"].map({
    "t": True,
    "f": False
})
# Convert room type
listings["room_type"] = (
    listings["room_type"]
    .str.strip()
    .str.title()
)
```

```python
print("Price datatype:", listings["price"].dtype)

print(
    "Missing prices:",
    listings["price"].isnull().sum()
)
```

```python
# ===============================================
# Data Validation
# ===============================================

print("Validating data...")

# Remove missing prices after conversion
listings = listings.dropna(subset=["price"])

# Remove zero or negative prices
listings = listings[listings["price"] > 0]

# Remove invalid latitude
listings = listings[
    (listings["latitude"] >= -90) &
    (listings["latitude"] <= 90)
]

# Remove invalid longitude
listings = listings[
    (listings["longitude"] >= -180) &
    (listings["longitude"] <= 180)
]
```

```python
# ===============================================
# Check Remaining Missing Values
# ===============================================

print("\nRemaining Missing Values")
print("-" * 40)

print(listings.isnull().sum().sort_values(ascending=False).head(20))
```

```python
# ===============================================
# Dataset Summary
# ===============================================

print("\nCleaned Dataset Summary")
print("-" * 40)

print("Listings :", listings.shape)
print("Calendar :", calendar.shape)
print("Reviews :", reviews.shape)
print("Neighbourhoods :", neighbourhoods.shape)
```

```python
print("Final listings:", listings.shape)
```

```python
# ===============================================
# Save Cleaned Data
# ===============================================

print("\nSaving cleaned datasets...")

listings.to_csv(
    "D:/Talent assessment/Data/cleaned/listings_clean.csv",
    index=False
)

calendar.to_csv(
    "D:/Talent assessment/Data/cleaned/calendar_clean.csv",
    index=False
)

reviews.to_csv(
    "D:/Talent assessment/Data/cleaned/reviews_clean.csv",
    index=False
)

neighbourhoods.to_csv(
    "D:/Talent assessment/Data/cleaned/neighbourhoods_clean.csv",
    index=False
)

print("\nData cleaning completed successfully!")
```
