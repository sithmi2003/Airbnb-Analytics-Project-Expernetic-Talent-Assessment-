```python
# ==========================================================
# statistical_analysis.py
# Airbnb Boston Statistical Analysis
# ==========================================================

import pandas as pd

# ==========================================================
# Load Dataset
# ==========================================================

print("Loading master dataset...\n")

df = pd.read_csv(
    "D:/Talent assessment/Data/cleaned/master_dataset.csv"
)

print("Dataset Loaded Successfully")
print("Shape:", df.shape)

# ==========================================================
# Dataset Information
# ==========================================================

print("\n" + "="*60)
print("DATASET INFORMATION")
print("="*60)

print(df.info())

# ==========================================================
# Descriptive Statistics
# ==========================================================

print("\n" + "="*60)
print("DESCRIPTIVE STATISTICS")
print("="*60)

numeric_columns = [
    "price",
    "accommodates",
    "bedrooms",
    "beds",
    "review_scores_rating",
    "occupancy_rate",
    "availability_rate",
    "total_reviews"
]

# Keep only columns that actually exist
numeric_columns = [col for col in numeric_columns if col in df.columns]

print(df[numeric_columns].describe())

# ==========================================================
# Missing Values
# ==========================================================

print("\n" + "="*60)
print("MISSING VALUES")
print("="*60)

print(df[numeric_columns].isnull().sum())

# ==========================================================
# Median
# ==========================================================

print("\n" + "="*60)
print("MEDIAN VALUES")
print("="*60)

print(df[numeric_columns].median())

# ==========================================================
# Standard Deviation
# ==========================================================

print("\n" + "="*60)
print("STANDARD DEVIATION")
print("="*60)

print(df[numeric_columns].std())

# ==========================================================
# Correlation Analysis
# ==========================================================

print("\n" + "="*60)
print("CORRELATION MATRIX")
print("="*60)

correlation = df[numeric_columns].corr(numeric_only=True)

print(correlation)

correlation.to_csv(
    "D:/Talent assessment/Data/cleaned/correlation_matrix.csv"
)

# ==========================================================
# Price by Room Type
# ==========================================================

print("\n" + "="*60)
print("PRICE BY ROOM TYPE")
print("="*60)

if "room_type" in df.columns:

    room_price = (
        df.groupby("room_type")["price"]
        .agg(["count", "mean", "median", "min", "max", "std"])
        .round(2)
    )

    print(room_price)

# ==========================================================
# Price by Superhost
# ==========================================================

print("\n" + "="*60)
print("PRICE BY SUPERHOST")
print("="*60)

if "host_is_superhost" in df.columns:

    superhost_price = (
        df.groupby("host_is_superhost")["price"]
        .agg(["count", "mean", "median"])
        .round(2)
    )

    print(superhost_price)

# ==========================================================
# Average Rating by Room Type
# ==========================================================

print("\n" + "="*60)
print("AVERAGE RATING BY ROOM TYPE")
print("="*60)

if (
    "room_type" in df.columns and
    "review_scores_rating" in df.columns
):

    rating = (
        df.groupby("room_type")["review_scores_rating"]
        .mean()
        .round(2)
    )

    print(rating)

# ==========================================================
# Occupancy by Room Type
# ==========================================================

print("\n" + "="*60)
print("OCCUPANCY RATE BY ROOM TYPE")
print("="*60)

if (
    "room_type" in df.columns and
    "occupancy_rate" in df.columns
):

    occupancy = (
        df.groupby("room_type")["occupancy_rate"]
        .mean()
        .round(2)
    )

    print(occupancy)

# ==========================================================
# Top 10 Most Expensive Neighbourhoods
# ==========================================================

print("\n" + "="*60)
print("TOP 10 MOST EXPENSIVE NEIGHBOURHOODS")
print("="*60)

if (
    "neighbourhood" in df.columns and
    "price" in df.columns
):

    top_neighbourhoods = (
        df.groupby("neighbourhood")["price"]
        .mean()
        .sort_values(ascending=False)
        .head(10)
        .round(2)
    )

    print(top_neighbourhoods)

# ==========================================================
# Top 10 Highest Rated Neighbourhoods
# ==========================================================

print("\n" + "="*60)
print("TOP 10 HIGHEST RATED NEIGHBOURHOODS")
print("="*60)

if (
    "neighbourhood" in df.columns and
    "review_scores_rating" in df.columns
):

    top_rating = (
        df.groupby("neighbourhood")["review_scores_rating"]
        .mean()
        .sort_values(ascending=False)
        .head(10)
        .round(2)
    )

    print(top_rating)

# ==========================================================
# Summary
# ==========================================================

print("\n" + "="*60)
print("SUMMARY")
print("="*60)

print(f"Total Listings           : {df['id'].nunique()}")

if "price" in df.columns:
    print(f"Average Price            : ${df['price'].mean():.2f}")

if "review_scores_rating" in df.columns:
    print(f"Average Rating           : {df['review_scores_rating'].mean():.2f}")

if "occupancy_rate" in df.columns:
    print(f"Average Occupancy Rate   : {df['occupancy_rate'].mean():.2f}%")

if "availability_rate" in df.columns:
    print(f"Average Availability     : {df['availability_rate'].mean():.2f}%")

print("\nStatistical analysis completed successfully.")
```
