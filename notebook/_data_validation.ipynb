```python
# ===============================================
# Airbnb Singapore Data Validation
# ===============================================

import pandas as pd

# ===============================================
# Load cleaned datasets
# ===============================================

print("Loading cleaned datasets...\n")

listings = pd.read_csv("D:/Talent assessment/Data/cleaned/listings_clean.csv")
calendar = pd.read_csv("D:/Talent assessment/Data/cleaned/calendar_clean.csv")
reviews = pd.read_csv("D:/Talent assessment/Data/cleaned/reviews_clean.csv")
neighbourhoods = pd.read_csv("D:/Talent assessment/Data/cleaned/neighbourhoods_clean.csv")

# ===============================================
# Dataset Shape
# ===============================================

print("=" * 60)
print("DATASET SHAPES")
print("=" * 60)

print(f"Listings       : {listings.shape}")
print(f"Calendar       : {calendar.shape}")
print(f"Reviews        : {reviews.shape}")
print(f"Neighbourhoods : {neighbourhoods.shape}")

# ===============================================
# Missing Values
# ===============================================

print("\n" + "=" * 60)
print("MISSING VALUES")
print("=" * 60)

print("\nListings")
print(listings.isnull().sum().sort_values(ascending=False).head(15))

print("\nCalendar")
print(calendar.isnull().sum())

print("\nReviews")
print(reviews.isnull().sum())

print("\nNeighbourhoods")
print(neighbourhoods.isnull().sum())

# ===============================================
# Duplicate Records
# ===============================================

print("\n" + "=" * 60)
print("DUPLICATE RECORDS")
print("=" * 60)

print("Listings :", listings.duplicated().sum())
print("Calendar :", calendar.duplicated().sum())
print("Reviews :", reviews.duplicated().sum())
print("Neighbourhoods :", neighbourhoods.duplicated().sum())

# ===============================================
# Data Types
# ===============================================

print("\n" + "=" * 60)
print("DATA TYPES")
print("=" * 60)

print("\nListings")
print(listings.dtypes)

print("\nCalendar")
print(calendar.dtypes)

print("\nReviews")
print(reviews.dtypes)

# ===============================================
# Price Validation
# ===============================================

print("\n" + "=" * 60)
print("PRICE VALIDATION")
print("=" * 60)

print("Missing Prices :", listings["price"].isnull().sum())
print("Negative Prices :", (listings["price"] < 0).sum())
print("Zero Prices :", (listings["price"] == 0).sum())

print("\nPrice Summary")
print(listings["price"].describe())

# ===============================================
# Latitude Validation
# ===============================================

print("\n" + "=" * 60)
print("LATITUDE VALIDATION")
print("=" * 60)

invalid_latitude = listings[
    (listings["latitude"] < -90) |
    (listings["latitude"] > 90)
]

print("Invalid Latitude Records :", len(invalid_latitude))

# ===============================================
# Longitude Validation
# ===============================================

print("\n" + "=" * 60)
print("LONGITUDE VALIDATION")
print("=" * 60)

invalid_longitude = listings[
    (listings["longitude"] < -180) |
    (listings["longitude"] > 180)
]

print("Invalid Longitude Records :", len(invalid_longitude))

# ===============================================
# Availability Validation
# ===============================================

print("\n" + "=" * 60)
print("AVAILABILITY VALIDATION")
print("=" * 60)

print(calendar["available"].value_counts(dropna=False))

# ===============================================
# Room Type Validation
# ===============================================

print("\n" + "=" * 60)
print("ROOM TYPE VALIDATION")
print("=" * 60)

print(listings["room_type"].value_counts())

# ===============================================
# Review Score Validation
# ===============================================

review_columns = [
    "review_scores_rating",
    "review_scores_accuracy",
    "review_scores_cleanliness",
    "review_scores_checkin",
    "review_scores_communication",
    "review_scores_location",
    "review_scores_value"
]

print("\n" + "=" * 60)
print("REVIEW SCORE VALIDATION")
print("=" * 60)

for column in review_columns:
    print(f"{column}:")
    print(listings[column].describe())
    print()

# ===============================================
# Bedrooms Validation
# ===============================================

print("=" * 60)
print("BEDROOM VALIDATION")
print("=" * 60)

print("Missing Bedrooms :", listings["bedrooms"].isnull().sum())
print("Minimum Bedrooms :", listings["bedrooms"].min())
print("Maximum Bedrooms :", listings["bedrooms"].max())

# ===============================================
# Beds Validation
# ===============================================

print("\n" + "=" * 60)
print("BEDS VALIDATION")
print("=" * 60)

print("Missing Beds :", listings["beds"].isnull().sum())
print("Minimum Beds :", listings["beds"].min())
print("Maximum Beds :", listings["beds"].max())

# ===============================================
# Final Validation Summary
# ===============================================

print("\n" + "=" * 60)
print("VALIDATION SUMMARY")
print("=" * 60)

print("✓ Missing values checked")
print("✓ Duplicate records checked")
print("✓ Price values validated")
print("✓ Latitude validated")
print("✓ Longitude validated")
print("✓ Data types verified")
print("✓ Room types verified")
print("✓ Availability verified")
print("✓ Review scores validated")

print("\nData validation completed successfully.")
```
